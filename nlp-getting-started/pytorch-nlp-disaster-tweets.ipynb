# Data cleaning

```python
def clean_html(text, show_example=True):
    if type(text) is not str:
        return text
    
    return re.sub(re.compile('<.*?>'), '', text)


def remove_url(text):
    if type(text) is not str:
        return text    
    
    url_regexp = r"https?:\/\/t.co\/[A-Za-z0-9]+"
    
    return re.sub(url_regexp, "", text)


def url_part_unquote(s):
    if type(s) is not str:
        return s
    return unquote(s)


def html_unescape(text):
    if type(text) is not str:
        return text        
    return html.unescape(text)

def replace_and(text):
    if type(text) is not str:
        return text
    return text.replace('&', ' and ')

# https://stackoverflow.com/a/47091370
def decontracted(text):
    if type(text) is not str:
        return text        
    
    # specific
    text = re.sub(r"won\'t", "will not", text)
    text = re.sub(r"can\'t", "can not", text)

    # general
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"([^-\s])\'re", r"\1 are", text)
    text = re.sub(r"([^-\s])\'s", r"\1 is", text)
    text = re.sub(r"([^-\s])\'d", r"\1 would", text)
    text = re.sub(r"([^-\s])\'ll", r"\1 will", text)
    text = re.sub(r"([^-\s])\'t", r"\1 not", text)
    text = re.sub(r"([^-\s])\'ve", r"\1 have", text)
    text = re.sub(r"([^-\s])\'m", r"\1 am", text)
    return text


def emoji_escape(text):
    def unicode_escape(chars, data_dict):
        return ' %s ' % emoji.demojize(chars)        
    
    if type(text) is not str:
        return text  
    
    return emoji.replace_emoji(text, replace=unicode_escape)


def remove_usernames(text):
    usernames_regexp = r'\@.*?\s'
    if type(text) is not str:
        return text      
    return re.sub(usernames_regexp, r' ', text)


def remove_punct(text):
    if type(text) is not str:
        return text      
    return text.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))


def utf8_only(text):
    utf8regexp = r'[^\x00-\x7f]'
    if type(text) is not str:
        return text      
    return re.sub(utf8regexp,r'', text)


def utf8_only(text):
    utf8regexp = r'[^\x00-\x7f]'
    if type(text) is not str:
        return text      
    return re.sub(utf8regexp,r'', text)


def remove_numbers(text):
    if type(text) is not str:
        return text    
    return re.sub(r'[0-9]+', '', text)


def replace_spaces(text):
    if type(text) is not str:
        return text
    result = (' '.join(text.split())).strip()
    return ''.join(result.splitlines())



data_cleaning_settings = {
    'html_tags': {
        'fn': clean_html,
        'example': '<span>text</span>'
    },
    'url': {
        'fn': remove_url,
        'example': 'my link http://t.co/Rs22LJ4qFp some text'
    },
    'url_part_unquote': {
        'fn': url_part_unquote,
        'example': 'violent%20storm'
    },
    'html_unescape': {
        'fn': html_unescape,
        'example': '&pound;682m'
    },
    'replace_and': {
        'fn': replace_and,
        'example': 'I am a #king & lion'
    },
    'decontraction': {
        'fn': decontracted,
        'example': "Hey I'm Yann, how're you and how's it going ? That's interesting: I'd love to hear more about it."
    },
    'emoji': {
        'fn': emoji_escape,
        'example': 'Python is 👍'
    },
    'usernames': {
        'fn': remove_usernames,
        'example': 'my username is @someuserasd. Find me'
    },
    'punctuation': {
        'fn': remove_punct,
        'example': 'I am a #king: lion'
    },
    'utf8': {
        'fn': utf8_only,
        'example': 'JetÛªs Debris Could Have Floated to RÌunion'
    },
    'remove_numbers': {
        'fn': remove_numbers,
        'example': 'test 8850 test 5568 test'
    },    
    'spaces': {
        'fn': replace_spaces,
        'example': 'I\nam a  king \n lion '
    }    
}

cleaning_steps = [
    'html_tags',
    'url', 
    'url_part_unquote',
    'html_unescape', 
    'replace_and', 
    'decontraction', 
    'emoji', 
    'usernames',
    'punctuation',
    'utf8',
    'remove_numbers',
    'spaces'
]

def data_processing(
    series,
    processing_settings,
    processing_steps,
    
    show_example=True,
    show_stats=True,
    show_real_example=True,
    ):
    
    result = series
    
    for processing_step_tag in processing_steps:
        processing_step = processing_settings[processing_step_tag]
        
        if show_example:
            example = processing_step['example']
            fn = processing_step['fn']
            print('%s\nexample: %s -> %s ' % (processing_step_tag, example, fn(example) )  )        

        new_result = result.apply(lambda x: fn(x)) 
        updated = (result != new_result) & (result.notna())
        if show_stats:
            print('cleaned lines: %s ' % updated.sum())
        if show_real_example and updated.sum() > 0:
            print('real example: ', result[updated].values[0], ' -> ' ,new_result[updated].values[0])

        result = new_result
        print()   
    
    return result
    
df_train['keyword_cleaned'] = data_processing(df_train['keyword'], data_cleaning_settings, cleaning_steps)
print(df_train[df_train['keyword_cleaned'].notna()]['keyword_cleaned'])
df_train['text_cleaned'] = data_processing(df_train['text'], data_cleaning_settings, cleaning_steps)
print(df_train[df_train['text_cleaned'].notna()]['text_cleaned'])
```

```python
def singularization(text):
    if type(text) is not str:
        return text 

    blob = TextBlob(text)
    r = text
    for word, tag in blob.tags:
        if tag.startswith('NN'):
            word_sing = word.singularize()
            r = r.replace(word, word_sing)
    return r

spell = SpellChecker()
def correct_spellings(text):
    if type(text) is not str:
        return text
    

    corrected_text = []
    misspelled_words = spell.unknown(text.split())

    for word in text.split():
        if (word in misspelled_words) and (spell.correction(word) is not None):
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)

data_lemmatizations_settings = {
    'singularization': {
        'fn': singularization,
        'example': 'nuclear disasters'
    },
    'spelling': {
        'fn': correct_spellings,
        'example': 'corect me plese'
    },   
}

lemmatizations_steps = [
    'singularization',
    'spelling',
]

df_train['keyword_transformed'] = data_processing(df_train['keyword_cleaned'], data_lemmatizations_settings, lemmatizations_steps)
print(df_train[df_train['keyword_transformed'].notna()]['keyword_transformed'])
```

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch import nn
from torch import optim
import torch


# train_dataloader = DataLoader(training_data, batch_size=64, shuffle=True)
# test_dataloader = DataLoader(test_data, batch_size=64, shuffle=True)
```

```python
class CustomTextDataset(Dataset):
    def __init__(self, input_ids, labels):
        self.input_ids = torch.tensor(input_ids)
        self.labels = torch.tensor(labels)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        input_ids = self.input_ids[idx]
        labels = self.labels[idx]
        sample = {"input_ids": input_ids, "labels": labels}
        return sample
```

```python
data_tr = CustomTextDataset(inputs['input_ids'][:split_int], df_train['target'].values[:split_int])
data_vl = CustomTextDataset(inputs['input_ids'][split_int:], df_train['target'].values[split_int:])
```

```python
dataloader_tr = DataLoader(data_tr, batch_size=64, shuffle=True)
dataloader_vl = DataLoader(data_vl, batch_size=64, shuffle=True)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters())

device = 'cpu'
```

```python
class LSTMModel(nn.Module):
    def __init__(self):
        super(LSTMModel, self).__init__()
        self.embedding = nn.Embedding(30000, 2000)
        self.input_size = 2000
        self.hidden_size = 1600
        self.bidirectional = True
        self.hidden_size_fc = int(self.hidden_size * (2 if self.bidirectional else 1))
#         self.lstm = nn.LSTM(input_size=1, hidden_size=10, num_layers=2, batch_first=True, bidirectional=True)
        self.lstm = nn.LSTM(
            input_size=self.input_size, 
            hidden_size=self.hidden_size, 
            num_layers=2, 
            batch_first=True, 
            dropout=0.1, 
            bidirectional = self.bidirectional
        )
#         self.relu = nn.ReLU()
        self.linear = nn.Linear(self.hidden_size_fc, 1)
        self.dropout = nn.Dropout(0.1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
#         print('0', x.shape)
        x = self.embedding(x)
        x = self.dropout(x)
#         print('1', x.shape)
        out, (hn, cn) = self.lstm(x)
#         print(out.shape, hn.shape, cn.shape)
#         hn = hn.view(-1, 10) #reshaping the data for Dense layer next
#         print('hn', hn.shape)

#         out_2 = out.squeeze()[-1, :]
        out_2 = out[:, -1, :]
        x = out_2
#         x = self.dropout(out_2)
#         print('out_2', out_2.shape)
#         x = self.relu(out_2)
        x = self.linear(x)
#         x = self.linear(hn)
#         print(x.shape)
        x = self.sigmoid(x)
#         print(x.shape)
        return x
```
